```python
over_50= heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            figsize=(10,6));

```

```python
#oo method
```

```python
fix,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x='age',
            y="chol",
            c="target",
            ax=ax);

#change the x axis limit
ax.set_xlim([45,100])
```

```python
fig,ax =plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);

#customized

ax.set(title ="heart dieases and chol level",
      xlabel ="Age",
      ylabel ="cholestrol");

#add legend
ax.legend(*scatter.legend_elements(),title ="Target");

#add horizonatal line

ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
over_50
```

```python
#subplot of chol ,age,thalach

fig,(ax0,ax1) =plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10))

#add data axes
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"]);
ax0.set(title ="Heart disease and cholestrol level",
       xlabel ="Age",
       ylabel="cholestrol");

#add legend

ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");


scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"]);
ax1.set(title ="Heart disease and thalach level",
       xlabel ="Age",
       ylabel="thalach(Max Heart Rate)");

#add legend

ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");


```

```python

#sharing the same values
fig,(ax0,ax1) =plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                           sharex=True)

#add data axes
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"]);
ax0.set(title ="Heart disease and cholestrol level",
       ylabel="cholestrol");

#add legend

ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");


scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"]);
ax1.set(title ="Heart disease and thalach level",
       xlabel ="Age",
       ylabel="thalach(Max Heart Rate)");

#add legend

ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

#add title figure

fig.suptitle("Heart Disease Analysis",fontsize =16,fontweight ="bold");
```

```python
#sharing the same values
fig,(ax0,ax1) =plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                           sharex=True)

#add data axes
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"],
                   cmap ="winter");
ax0.set(title ="Heart disease and cholestrol level",
       ylabel="cholestrol");

ax0.set_xlim([50,80])

#add legend

ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");


scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"],
                   cmap="winter");
ax1.set(title ="Heart disease and thalach level",
       xlabel ="Age",
       ylabel="thalach(Max Heart Rate)");

#add legend

ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
#add title figure


fig.suptitle("Heart Disease Analysis",fontsize =16,fontweight ="bold");
```
