```python
%matplotlib inline
import matplotlib.pyplot as plt
x =[1,2,3,4]
y =[11,22,33,44]

fig,ax=plt.subplots(figsize=(10,10))

ax.plot(x,y)
ax.set(title ="simple plot",
      xlabel ="x -axis",
      ylabel ="y -axis")

fig.savefig("sample-pot.png")
```
